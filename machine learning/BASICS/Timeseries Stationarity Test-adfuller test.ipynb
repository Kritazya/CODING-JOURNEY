# Augmented Dickey Fuller test(ADF Test):
* is a common statistical test used to test whether a given Time series is stationary or not. It is one of the

```python
import numpy as np
data = [3,4,5,4.4,6,5.5,7,8,9]
```

```python
import matplotlib.pyplot as plt
plt.plot(data, 'o-', color = 'purple', ms = 10)
plt.xlabel("Time")
plt.ylabel("growth")
plt.show()
```

```python
from statsmodels.tsa.stattools import adfuller
pvalue = adfuller(x = data)[1]
```

```python
# Null hypothesis
alternative_hypothesis = "The data is stationary"
null_hypothesis = "The data is not stationary"
if pvalue < 0.05:
  print("We reject null hypothesis")
else:
  print("We fail to reject null hypothesis. Which means data is not stationary")
```
