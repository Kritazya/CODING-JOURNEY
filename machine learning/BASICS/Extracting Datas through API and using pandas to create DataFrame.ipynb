```python
import requests
import pandas as pd
```

```python
response = requests.get('https://newsapi.org/v2/everything?q=tesla&from=2023-05-22&sortBy=publishedAt&apiKey=785ade9cb6ec4835950ee575d9f33121').json()
```

```python
author = []
title = []
description = []
publishedAt = []
content = []
for i in response['articles']:
  author.append(i['author'])
  title.append(i['title'])
  description.append(i['description'])
  publishedAt.append(i['publishedAt'])
  content.append(i['content'])
d = {'author': author, 'title': title, 'description': description, 'publishedAt': publishedAt, 'content': content}
df = pd.DataFrame(d)
df.head()
```

```python
df1 = pd.DataFrame(response['articles'])
df1.drop(['source','url', 'urlToImage'], 1, inplace = True)
df1.head()
```
