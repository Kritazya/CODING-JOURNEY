```python
import numpy as np
import matplotlib.pyplot as plt
noise = np.random.normal(loc=0, scale=0.5, size = 1000)
plt.figure(figsize= (12,4))
plt.plot(noise)
```

```python
from statsmodels.tsa import stattools
autocorrelations = stattools.acf(noise, nlags=40, fft=False)
fig = plt.figure(figsize=(10,2.7))
# plt.stem(autocorrelations, use_line_collection = True)
plt.stem(autocorrelations)
plt.title("Autocorrelation of a White Noise Series")
plt.xticks(range(0,49,12), fontsize = 13)
plt.grid()
plt.show()
```
