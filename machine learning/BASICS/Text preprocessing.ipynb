# TOKENIZATION



```python
import nltk
msg = "Hi, would you like to go to dinner with me?"
nltk.download('punkt')
nltk.word_tokenize(msg)
```

```python
msg1 = "I love football. I like cricket too."
nltk.sent_tokenize(msg1)
```

#Stemming
* Base word or root word

```python
import nltk
nltk.download('punkt')
words = ['change', 'changing', 'changes']
stemmer = nltk.stem.PorterStemmer()
[stemmer.stem(word) for word in words]
```

# Lemmatization

```python
import nltk
nltk.download('punkt')
nltk.download('wordnet')
words = ['change', 'changing', 'changes', 'changer', 'changed']
lemmatizer = nltk.stem.WordNetLemmatizer()
[lemmatizer.lemmatize(word) for word in words]
```

# BAG of WORDS / Count Vectorizer

```python
from sklearn.feature_extraction.text import CountVectorizer
corpus = ['he is a good good boy', 'she is a good girl', 'boy and girl are good']
cv = CountVectorizer(stop_words = 'english')
```

```python
cv.fit_transform(corpus)
```

```python
sparse = cv.fit_transform(corpus)
sparse.toarray()
```

```python
corpus
```

```python
cv.get_feature_names_out()
```

```python
import pandas as pd
pd.DataFrame(sparse.toarray(), columns=cv.get_feature_names_out())
```

# TF - IDF  Vectorizer

```python
from sklearn.feature_extraction.text import TfidfVectorizer
corpus = ['he is a good good boy', 'she is a good girl', 'boy and girl are good']
vectorizer = TfidfVectorizer(stop_words = 'english')
sparse_vector = vectorizer.fit_transform(corpus)
```

```python
sparse_vector.toarray()
```

```python
vectorizer.get_feature_names_out()
```

```python
pd.DataFrame(sparse_vector.toarray(), columns=vectorizer.get_feature_names_out())
```
