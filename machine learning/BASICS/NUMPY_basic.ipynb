# List supports multiple datatype while numpy array is homogenous

```python
list1 = [1,1,1]
x = np.array([list1], dtype= np.int0)
x
```

```python
import random
random.seed(80)
data = [random.randint(1,10) for item in range(100000)]
```

```python
print(data)
```

```python
data_array = np.array(data)
data_array
```

```python
%%time
sq = [item ** 2 for item in data]
```

```python
%%time
sq = data_array**2
```
