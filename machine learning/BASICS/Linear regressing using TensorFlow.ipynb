```python
import numpy as np
import tensorflow as tf
```

```python
xi = np.array([-1.0, 2.0, 5.0, 7.0])
yi = np.array([-1.0, 5.0, 11.0, 15.0])
```

```python
model = tf.keras.Sequential()
```

```python
model.add(tf.keras.layers.Dense(1, input_shape = [1]))
```

# After making model we must compile it

```python
model.compile(optimizer = 'sgd', loss = 'mean_squared_error')
```

```python
history = model.fit(xi , yi, epochs = 100)
```

```python
import matplotlib.pyplot as plt
loss = history.history['loss']
epoch = history.epoch
```

```python
model_checkpoint = tf.keras.callbacks.ModelCheckpoint('model.h5', monitor = 'loss', save_best_only = True)
```

```python
early_stopping = tf.keras.callbacks.EarlyStopping(monitor = 'loss', patience=5)
```

```python
callbacks = [model_checkpoint, early_stopping]
```

```python
model = tf.keras.Sequential()
model.add(tf.keras.layers.Dense(1, input_shape = [1]))
model.compile(optimizer = 'sgd', loss = 'mean_squared_error')
```

```python
model.fit(xi,yi, epochs = 3000, callbacks = callbacks) #
```
