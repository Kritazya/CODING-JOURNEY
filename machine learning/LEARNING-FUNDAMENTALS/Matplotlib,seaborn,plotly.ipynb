# BAR-PLOT

```python
player_name = ['Kushal','Aasif', 'Gulsan', 'Bhim']
scores= [1,8,67,36]
```

```python
plt.bar(x = player_name, height = scores)
```

```python
players_name = ['Kushal','Aasif', 'Gulsan', 'Bhim','Sandeep']
runs= [1,8,16,8,1]
plt.bar(x = players_name, height = runs, color=["red", "blue", "black","yellow",'pink'])
```

# SEA-BORN

```python
sns.barplot(x =player_name, y= scores)
plt.title("Players and their scores")
plt.ylabel("Scores")
plt.show()
```

#Plotly

```python
fig = px.bar(x = player_name, y=scores)
fig.update_layout(title = "Players and their scores",
                  width = 700,
                  height=400,
                  xaxis_title = "Players Name",
                  yaxis_title= "Runs")
```

# Scatter-Plot

```python
import numpy as np
age = np.array([5,6,7,8.2,9,9.3])
weight = np.array([20,22,28,28,25,28])
```

```python
plt.figure(figsize=[5,3])
plt.scatter(x=age, y=weight)
plt.title("Relationship between age and weight")
plt.xlabel("age")
plt.ylabel("weight")
plt.show()
```

```python
plt.figure(figsize=[5,3])
sns.scatterplot(x = age, y=weight, marker = 'x', color = 'red', s=100, size=20) #SIZE=????
plt.title("Relationship between age and weight")
plt.xlabel("age")
plt.ylabel("weight")
plt.show()
```

# PIE-CHART

```python
player_name = ['Kushal','Aasif', 'Gulsan', 'Bhim']
scores= [1,8,67,36]
fig = px.pie(names = player_name, values=scores)
fig.update_layout(width = 700, height = 400)
```

```python
player_name = ['Kushal','Aasif', 'Gulsan', 'Bhim']
scores= [1,8,67,36]
fig = px.pie(names = player_name, values=scores)
fig.update_layout(width = 700, height = 400)
fig.update_traces(pull = 0.1)
```
