```python
import pandas as pd
df = pd.read_csv('/content/classified_data.csv')
```

```python
df.head()
```

```python
df.drop(columns= ['Unnamed: 0'], inplace = True)
```

```python
df.head()
```

```python
df.describe()
```

```python
df.isna().sum()
```

```python
X = df.iloc[:, :-1]
y = df.iloc[:, -1]
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
scaled_x = sc.fit_transform(X)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
nn = range(1,30)
```

```python
from sklearn.model_selection import cross_validate
error = []
for k in nn:
  model = KNeighborsClassifier(n_neighbors = k)
  scores = cross_validate(estimator = model, X = scaled_x ,y=y, cv=10, scoring = 'f1')
  error.append(1-scores['test_score'].mean())

```

```python
import plotly.express as px
fig = px.line(x = nn, y=error)
fig.update_layout(width = 600, height = 500)
```

```python
k = 14
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(scaled_x , y, test_size = 0.2, random_state=40)
```

```python
model = KNeighborsClassifier(n_neighbors=k)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
from sklearn.metrics import f1_score
f1 = f1_score(y_test, y_pred)
print(f1)
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
confusion_matrix(y_test, y_pred)
```
