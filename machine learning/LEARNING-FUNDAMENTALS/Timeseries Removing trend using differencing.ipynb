```python
import pandas as pd
df = pd.read_csv('/content/microsoft_data.csv')
df.head()
```

```python
import matplotlib.pyplot as plt
plt.plot(df.Close)
```

```python
# adfuller test to check if  there is trend or not
from statsmodels.tsa.stattools import adfuller
pvalue = adfuller(x= df. Close)[1]
null_hypothesis = "The data is not stationary."
alternative_hypothesis = "Data is stationary"
if pvalue < 0.05:
  print("We reiect null hypothesis")
else:
  print("We fail to reject null hypothesis. Which means data 1s not stationary")
```
