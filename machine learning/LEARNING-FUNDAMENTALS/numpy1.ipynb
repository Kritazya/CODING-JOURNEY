# Normal and Standard Normal distribution

Normal
* mean = $\mu$
* variance = $\sigma^2$

Standard Normal Distribution
* mean = 0
* variance = 1


```python
import numpy as np
dat = np.random.normal(loc = 5, scale = 1, size = 2000) # loc = mean, scale = std, size = no.of data points
```

```python
import seaborn as sns
sns.histplot(dat, kde=True)
```

```python
dat = np.random.standard_normal(size=2000)
sns.histplot(dat, kde = True)
```

```python
sns.boxplot(y = dat)
```
