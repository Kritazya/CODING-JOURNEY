# Vertical and Horizontal Edge Detection


```python
!wget https://a-z-animals.com/media/2021/11/Zebra-in-field-e1653793342190-1024x535.jpg
```

```python
import cv2
import matplotlib.pyplot as plt
image = cv2.imread('/content/Zebra-in-field-e1653793342190-1024x535.jpg', 0)
plt.imshow(image, cmap = 'gray')
```

# Detect Vertical Edge

```python
import numpy as np
kernel0 = np.array([[-1, 0, 1], [-1, 0, 1], [-1, 0, 1]])
```

```python
# Same depth as the original image => ddepth = -1
vertical_edge = cv2.filter2D(image, kernel=kernel0, ddepth = -1)
```

```python
plt.imshow(vertical_edge, cmap='gray')
```

# Horizontal Edge Detection

```python
kernel1 = np.array([[1, 1, 1], [0, 0, 0], [-1, -1, -1]])
```

```python
horizontal_edge = cv2.filter2D(image, kernel = kernel1, ddepth = -1)
```

```python
plt.imshow(horizontal_edge, cmap = 'gray')
```

```python
#SOMETHING IS FISHY
plt.subplots(1,2, figsize = (15,4) )
```

```python
fig, axes = plt.subplots(1,2, figsize = (15,4) )
ax = axes[0]
bx = axes[1]
ax.imshow(vertical_edge, cmap = 'gray')
ax.set_title("Image after applying vertical edge detection filter")
bx.imshow(horizontal_edge, cmap = 'gray')
bx.set_title("Image after applying horizontal edge detection filter")
fig.show()
```

# Image Sharpening => Highlights edges and fine details of the image => High pass filter

```python
import cv2
import matplotlib.pyplot as plt
image = cv2.imread('/content/Zebra-in-field-e1653793342190-1024x535.jpg', 0)
plt.imshow(image, cmap = 'gray')
```

```python
kernel = np.array([[0,-1,0], [-1,5,-1], [0,-1,0]])  #More sharpening happens if we change the value of 5
sharpened_image = cv2.filter2D(image, kernel = kernel, ddepth=-1)
plt.imshow(sharpened_image, cmap='gray')
```

# Blur Image

```python
blurred_image= cv2.blur(image, ksize = (11,11))
plt.imshow(blurred_image, cmap = 'gray')
```
