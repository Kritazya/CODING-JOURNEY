```python
df = pd.DataFrame()
for i in range(5):
  headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 6.3; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.162 Safari/537.36'
  } # Use this when it shows error while not using this!ðŸ˜€
  webpage = requests.get(f'https://gomovies.sx/genre/horror?page={i+1}',headers = headers).text
  soup = BeautifulSoup(webpage, 'lxml')
  for j in range(32):
    b = soup.find_all('a', class_='film-poster-ahref flw-item-tip')[j]['href']
    url = 'https://gomovies.sx' + b
    webpage1 = requests.get(url).text
    soup1 = BeautifulSoup(webpage1, 'lxml')
    rating = soup1.find('button', class_='btn btn-sm btn-radius btn-warning btn-imdb').text
    if rating.strip('IMDB: ') == 'N/A':
      pass
    else:
      name = soup1.find('h2',class_='heading-name').text
      rating = float(rating.strip('IMDB: '))
      df = df.append({'title': name, 'Rating': rating}, ignore_index = True)
df = df.sort_values('Rating', ascending = False)
df = df.reset_index(drop=True)
print(df.to_string())
```
